````python
year = 2005
month = 5

with engine.connect() as conn:
    # Run SQL query
    result = conn.execute(text(f"""
        SELECT *
        FROM rental
        WHERE DATE_FORMAT(CONVERT(rental_date, DATE), '%Y') = {year} AND DATE_FORMAT(CONVERT(rental_date, DATE), '%m') = {month};
    """))

    # Fetch all rows
    tuple_result = result.fetchall()

# Retrieve column names from the result set
columns = result.keys()

# Convert fetched rows to a Pandas DataFrame
df_rental = pd.DataFrame(tuple_result, columns=columns)
df_rental.head()
````

````python
def rentals_month(month:int, year:int):

    if isinstance(month, int) and isinstance(year, int):
        with engine.connect() as conn:
            # Run SQL query
            result = conn.execute(text(f"""
                SELECT *
                FROM rental
                WHERE DATE_FORMAT(CONVERT(rental_date, DATE), '%Y') = {year} AND DATE_FORMAT(CONVERT(rental_date, DATE), '%m') = {month};
            """))
        
            # Fetch all rows
            tuple_result = result.fetchall()
        
        # Retrieve column names from the result set
        columns = result.keys()
        
        # Convert fetched rows to a Pandas DataFrame
        df_rental = pd.DataFrame(tuple_result, columns=columns)
        return df_rental
````

````python
def rental_count_month(df, month:int, year:int):
    df_rcm = df.copy()
    df_rcm['year'] = df_rental['rental_date'].dt.year.astype(int)
    df_rcm['month'] = df_rental['rental_date'].dt.month.astype(int)
    #print(df_rcm)
    
    #df_rcm = df_rcm[(df_rcm['year']== year) & (df_rcm['month']==month)]
    #print(df_rcm['year'])
    
    df_rcm_out = df_rcm.groupby(["customer_id", "year", "month"])["rental_id"].count().sort_values(ascending=False).reset_index()
    #print(df_rcm_out)
    
    df_rcm_out.columns = ['customer_id', 'year', 'month', f'rentals_{month}_{year}']

    return df_rcm_out

````

Create a Python function called `compare_rentals` that takes two DataFrames as input containing the number of rentals made by each customer in different months and years. 
The function should return a combined DataFrame with a new 'difference' column, which is the difference between the number of rentals in the two months.

````python
rentals_col = [ele for ele in df_month_1.columns if re.match(r'rentals_\d{1}_\d{4}', ele)]
rentals_col
````

````python
def compare_rentals(df_1:pd.DataFrame, df_2:pd.DataFrame):
    df_c1 = df_1.copy()
    rentals_col_1 = [ele for ele in df_c1.columns if re.match(r'rentals_\d{1}_\d{4}', ele)]
    
    df_c2 = df_2.copy()
    rentals_col_2 = [ele for ele in df_c2.columns if re.match(r'rentals_\d{1}_\d{4}', ele)]
    
    df_merged = pd.merge(df_1, df_2[["customer_id", rentals_col_2[0]]], on='customer_id') # could youse another joining function with how

    df_merged['difference'] = (df_merged[rentals_col_1[0]] - df_merged[rentals_col_2[0]]).abs()

    return df_merged
````

````python
df_compare_rentals = compare_rentals(df_month_1, df_month_2)
df_compare_rentals.sort_values("difference", ascending=False)
````
